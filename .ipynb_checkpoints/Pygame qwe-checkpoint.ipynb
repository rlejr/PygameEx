```python

import pygame

# 전체적인 구성

# 1. 초기화 -> 변수 선언(변수를 사용할 준비)
pygame.init()

# 2. 게임 화면 설정 -> 크기 고정
size = [400, 900]
screen = pygame.display.set_mode(size)

title = "PyGameEx"

pygame.display.set_caption(title)

# 3. 게임 화면 내에서의 설정 -> 변수
clock = pygame.time.Clock()



color = (0, 0, 0)
black = (0, 0, 0)
custom_color = (63, 171, 126)

class SetImage:
    
    def __init__(self):
        self.x = 0
        self.y = 0
        
   
    def loadAndSetimage(self, imageRoot, scale):
        if imageRoot[-3:] =="png":
            self.image = pygame.image.load(imageRoot).convert_alpha()
        else:
            self.image = pygame.image.load(imageRoot)
            
        
        self.width, self.height = self.image.get_size()
        self.image = pygame.transform.scale(self.image, (round(self.width * scale), round(self.height * scale)))
        self.width, self.height = self.image.get_size()
    
    def show(self):
        screen.blit(self.image, (self.x , self.y))

                    
hero = SetImage();
hero.loadAndSetimage('C:/Users/jeong/Desktop/clone pygame/PygameEx/hero.png', 3/9)
hero.x = round(size[0] / 2) - round(hero.width / 2)
hero.y = size[1] - (hero.height + 50)
hero.move = 10
left_move = False
right_move = False
# 4. 메인이벤트
system_exit = 0
while(system_exit == 0):
    
    #  4-1. FPS(Frame Per Sec) 설정
    clock.tick(60)
    #  4-2. 입력(키보드, 마우스)의 감지
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            system_exit = 1
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                left_move = True
            if event.key == pygame.K_RIGHT:
                right_move = True
        
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_move = False
            if event.key == pygame.K_RIGHT:
                right_move = False
    
    
    #  4-3. 입력, 시간에 따른 변화
    if hero.x != 0 :
        if left_move == True:
            hero.x -= hero.move
    if hero.x != (size[0]-hero.width):
        if right_move == True:
            hero.x += hero.move
    #  4-4. 전사작업(그리기)
    screen.fill(color)
    hero.show()
    #  4-5. 업데이트
    pygame.display.flip()
    
# 5. 게임종료
pygame.quit()
```
